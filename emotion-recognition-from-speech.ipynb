```python
# ============================================================
# SPEECH EMOTION RECOGNITION - MODIFIED FULL PIPELINE
# RAVDESS DATASET
# Fixes prediction issue and improves training quality
# ============================================================

# =========================
# 1) SETUP
# =========================

import os
import json
import random
import warnings
warnings.filterwarnings("ignore")

os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import librosa

import tensorflow as tf
from tensorflow.keras import backend as K
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_recall_fscore_support
)
from sklearn.utils.class_weight import compute_class_weight

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

print("TensorFlow:", tf.__version__)

# =========================
# 2) DATASET PATH
# =========================

dataset_path = "/kaggle/input/datasets/uwrfkaggler/ravdess-emotional-speech-audio"

if not os.path.exists(dataset_path):
    raise FileNotFoundError(f"Dataset path not found: {dataset_path}")

print("Dataset path found.")

# =========================
# 3) EMOTION MAP
# =========================

emotion_dict = {
    "01": "neutral",
    "02": "calm",
    "03": "happy",
    "04": "sad",
    "05": "angry",
    "06": "fearful",
    "07": "disgust",
    "08": "surprised"
}

# =========================
# 4) COLLECT FILES
# =========================

file_paths = []
labels = []

for root, _, files in os.walk(dataset_path):
    for file in files:
        if file.lower().endswith(".wav"):
            parts = file.split("-")
            if len(parts) >= 3 and parts[2] in emotion_dict:
                file_paths.append(os.path.join(root, file))
                labels.append(emotion_dict[parts[2]])

df = pd.DataFrame({"path": file_paths, "label": labels})

print("Total audio files:", len(df))
print("\nClass distribution:")
print(df["label"].value_counts())

if len(df) == 0:
    raise ValueError("No WAV files found. Check the dataset path.")

# =========================
# 5) LABEL ENCODING
# =========================

le = LabelEncoder()
y_all = le.fit_transform(df["label"].values)
class_names = le.classes_
num_classes = len(class_names)

print("\nClasses:", list(class_names))
print("Num classes:", num_classes)

if num_classes < 2:
    raise ValueError("Need at least two classes.")

# =========================
# 6) TRAIN / VAL / TEST SPLIT
# =========================

train_files, temp_files, train_labels, temp_labels = train_test_split(
    df["path"].values,
    y_all,
    test_size=0.20,
    random_state=SEED,
    stratify=y_all
)

val_files, test_files, val_labels, test_labels = train_test_split(
    temp_files,
    temp_labels,
    test_size=0.50,
    random_state=SEED,
    stratify=temp_labels
)

print("\nSplit sizes:")
print("Train:", len(train_files))
print("Val:  ", len(val_files))
print("Test: ", len(test_files))

# =========================
# 7) AUDIO AUGMENTATION
# =========================

def augment_audio(audio, sr):
    choice = random.choice(["noise", "shift", "pitch", "stretch"])
    try:
        if choice == "noise":
            noise = 0.005 * np.random.randn(len(audio))
            return audio + noise

        if choice == "shift":
            shift = np.random.randint(-int(0.2 * sr), int(0.2 * sr))
            return np.roll(audio, shift)

        if choice == "pitch":
            steps = np.random.uniform(-2, 2)
            return librosa.effects.pitch_shift(audio, sr=sr, n_steps=steps)

        if choice == "stretch":
            rate = np.random.uniform(0.85, 1.15)
            return librosa.effects.time_stretch(audio, rate=rate)

    except Exception:
        return audio

    return audio

# =========================
# 8) FEATURE EXTRACTION
#    MFCC + Delta + Delta-Delta
# =========================

N_MFCC = 40
MAX_LEN = 173   # fixed time dimension

def pad_or_truncate(mat, max_len=MAX_LEN):
    if mat.shape[1] < max_len:
        pad_width = max_len - mat.shape[1]
        mat = np.pad(mat, ((0, 0), (0, pad_width)), mode="constant")
    else:
        mat = mat[:, :max_len]
    return mat

def load_audio(file_path):
    try:
        audio, sr = librosa.load(file_path, duration=3, offset=0.5)
        if audio is None or len(audio) == 0:
            return None, None
        return audio, sr
    except Exception:
        return None, None

def extract_features_from_audio(audio, sr):
    try:
        mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=N_MFCC)
        delta = librosa.feature.delta(mfcc)
        delta2 = librosa.feature.delta(mfcc, order=2)

        mfcc = pad_or_truncate(mfcc)
        delta = pad_or_truncate(delta)
        delta2 = pad_or_truncate(delta2)

        feat = np.stack([mfcc, delta, delta2], axis=-1).astype(np.float32)  # (40,173,3)
        return feat
    except Exception:
        return None

# =========================
# 9) BUILD DATASET
# =========================

def build_dataset(file_list, label_list, augment=False, show_progress=True):
    X, y = [], []

    for i, (fp, lbl) in enumerate(zip(file_list, label_list), start=1):
        audio, sr = load_audio(fp)
        if audio is None:
            continue

        feat = extract_features_from_audio(audio, sr)
        if feat is not None:
            X.append(feat)
            y.append(lbl)

        if augment:
            aug_audio = augment_audio(audio, sr)
            feat_aug = extract_features_from_audio(aug_audio, sr)
            if feat_aug is not None:
                X.append(feat_aug)
                y.append(lbl)

        if show_progress and i % 100 == 0:
            print(f"Processed {i}/{len(file_list)} files")

    return np.array(X, dtype=np.float32), np.array(y, dtype=np.int32)

print("\nExtracting training features...")
X_train, y_train = build_dataset(train_files, train_labels, augment=True)

print("\nExtracting validation features...")
X_val, y_val = build_dataset(val_files, val_labels, augment=False)

print("\nExtracting test features...")
X_test, y_test = build_dataset(test_files, test_labels, augment=False)

# =========================
# 10) SANITY CHECKS
# =========================

print("\nSanity checks:")
print("X_train:", X_train.shape)
print("X_val:  ", X_val.shape)
print("X_test: ", X_test.shape)

print("y_train:", y_train.shape)
print("y_val:  ", y_val.shape)
print("y_test: ", y_test.shape)

if len(X_train) == 0 or len(X_val) == 0 or len(X_test) == 0:
    raise ValueError("One of the splits has no extracted features.")

expected_shape = (N_MFCC, MAX_LEN, 3)
if X_train.shape[1:] != expected_shape:
    raise ValueError(f"Unexpected input shape. Expected {expected_shape}, got {X_train.shape[1:]}")

if np.isnan(X_train).any() or np.isnan(X_val).any() or np.isnan(X_test).any():
    raise ValueError("NaN found in extracted features.")

print("\nFeature extraction successful.")

# =========================
# 11) NORMALIZATION
# =========================

train_mean = X_train.mean(axis=(0, 1, 2), keepdims=True).astype(np.float32)
train_std = (X_train.std(axis=(0, 1, 2), keepdims=True) + 1e-8).astype(np.float32)

X_train = (X_train - train_mean) / train_std
X_val   = (X_val - train_mean) / train_std
X_test  = (X_test - train_mean) / train_std

print("\nNormalization done.")
print("Train mean:", float(X_train.mean()))
print("Train std :", float(X_train.std()))

# =========================
# 12) ONE-HOT ENCODING
# =========================

y_train_cat = to_categorical(y_train, num_classes=num_classes)
y_val_cat   = to_categorical(y_val, num_classes=num_classes)
y_test_cat  = to_categorical(y_test, num_classes=num_classes)

# =========================
# 13) CLASS BALANCING
# =========================

classes = np.unique(y_train)
class_weights_arr = compute_class_weight(
    class_weight="balanced",
    classes=classes,
    y=y_train
)
class_weight_dict = {int(c): float(w) for c, w in zip(classes, class_weights_arr)}
sample_weights = np.array([class_weight_dict[int(lbl)] for lbl in y_train], dtype=np.float32)

print("\nClass weights:")
print(class_weight_dict)

# =========================
# 14) ADVANCED LOSS: FOCAL LOSS
# =========================

def categorical_focal_loss(gamma=2.0, alpha=0.25, label_smoothing=0.05):
    def loss(y_true, y_pred):
        y_true = tf.cast(y_true, tf.float32)
        y_pred = tf.cast(y_pred, tf.float32)

        ncls = tf.cast(tf.shape(y_true)[-1], tf.float32)
        y_true = y_true * (1.0 - label_smoothing) + label_smoothing / ncls

        y_pred = tf.clip_by_value(y_pred, K.epsilon(), 1.0 - K.epsilon())
        ce = -y_true * tf.math.log(y_pred)
        weight = alpha * tf.pow(1.0 - y_pred, gamma)

        return tf.reduce_sum(weight * ce, axis=-1)
    return loss

# =========================
# 15) MODEL
# =========================

def build_cnn_model(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)

    x = layers.Conv2D(32, (3, 3), padding="same", use_bias=False)(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Dropout(0.20)(x)

    x = layers.Conv2D(64, (3, 3), padding="same", use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Dropout(0.25)(x)

    x = layers.Conv2D(128, (3, 3), padding="same", use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Dropout(0.30)(x)

    x = layers.Conv2D(256, (3, 3), padding="same", use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.GlobalAveragePooling2D()(x)

    x = layers.Dense(256, activation="relu", kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.Dropout(0.40)(x)

    x = layers.Dense(128, activation="relu", kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.Dropout(0.30)(x)

    outputs = layers.Dense(num_classes, activation="softmax")(x)

    return models.Model(inputs, outputs)

model = build_cnn_model(X_train.shape[1:], num_classes)

try:
    optimizer = tf.keras.optimizers.AdamW(learning_rate=1e-3, weight_decay=1e-5)
except Exception:
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)

model.compile(
    optimizer=optimizer,
    loss=categorical_focal_loss(gamma=2.0, alpha=0.25, label_smoothing=0.05),
    metrics=["accuracy"]
)

print("\nModel summary:")
model.summary()

# =========================
# 16) CALLBACKS
# =========================

callbacks = [
    EarlyStopping(monitor="val_loss", patience=12, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6, verbose=1),
    ModelCheckpoint("best_emotion_recognition_model.keras", monitor="val_accuracy", save_best_only=True, verbose=1)
]

# =========================
# 17) TRAIN
# =========================

history = model.fit(
    X_train,
    y_train_cat,
    validation_data=(X_val, y_val_cat),
    epochs=60,
    batch_size=32,
    sample_weight=sample_weights,
    callbacks=callbacks,
    verbose=1
)

# =========================
# 18) EVALUATION
# =========================

test_loss, test_acc = model.evaluate(X_test, y_test_cat, verbose=0)
print(f"\nTest Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")

y_prob = model.predict(X_test, verbose=0)
y_pred = np.argmax(y_prob, axis=1)
y_true = y_test

accuracy = accuracy_score(y_true, y_pred)
precision, recall, f1, _ = precision_recall_fscore_support(
    y_true, y_pred, average="macro", zero_division=0
)

print("\nFinal Metrics:")
print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1-score : {f1:.4f}")

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names, digits=4, zero_division=0))

# =========================
# 19) PLOTS
# =========================

plt.figure(figsize=(10, 4))
plt.plot(history.history["accuracy"], label="Train Accuracy")
plt.plot(history.history["val_accuracy"], label="Val Accuracy")
plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)

plt.figure(figsize=(10, 8))
sns.heatmap(cm_norm, annot=True, fmt=".2f", cmap="Greens",
            xticklabels=class_names, yticklabels=class_names)
plt.title("Normalized Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# =========================
# 20) SAVE MODEL + CLASS NAMES
# =========================

model.save("emotion_recognition_best.keras")

with open("label_classes.json", "w") as f:
    json.dump(class_names.tolist(), f)

np.save("train_mean.npy", train_mean)
np.save("train_std.npy", train_std)

print("\nSaved model as emotion_recognition_best.keras")
print("Saved label classes as label_classes.json")
print("Saved normalization stats as train_mean.npy and train_std.npy")

# =========================
# 21) FIXED PREDICTION FUNCTION
# =========================

def predict_emotion(audio_path):
    audio, sr = load_audio(audio_path)
    if audio is None:
        return "Audio load failed", 0.0

    feat = extract_features_from_audio(audio, sr)
    if feat is None:
        return "Feature extraction failed", 0.0

    # Normalize with training statistics
    feat = (feat - train_mean.squeeze()) / train_std.squeeze()

    # Add batch dimension
    feat = np.expand_dims(feat, axis=0).astype(np.float32)

    # Sanity check
    print("Prediction input shape:", feat.shape)

    prob = model.predict(feat, verbose=0)[0]
    pred_idx = int(np.argmax(prob))
    pred_label = le.inverse_transform([pred_idx])[0]
    confidence = float(np.max(prob))

    return pred_label, confidence

# =========================
# 22) TEST SINGLE PREDICTION
# =========================

sample_path = test_files[0]
pred_label, pred_conf = predict_emotion(sample_path)

print("\nSample Prediction Test:")
print("File:", sample_path)
print("Predicted Emotion:", pred_label)
print("Confidence:", round(pred_conf, 4))

# =========================
# 23) FINAL SANITY CHECK
# =========================

print("\n===================================================")
print("PIPELINE COMPLETED SUCCESSFULLY")
print("Model trained, evaluated, and prediction fixed.")
print("===================================================")
```
